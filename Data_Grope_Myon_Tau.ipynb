```python
def particleID_two(Ctrk,Sump,Ecal,Hcal):
    tau_count=0
    e_count=0
    had_count=0
    m_count=0
    for i in range(len(Ctrk)):
        if (t_cuts[0]<=Ctrk[i]<=t_cuts[1]) and (t_cuts[3]<=Sump[i]<=t_cuts[4]) and (t_cuts[5]<=Ecal[i]<=t_cuts[6]) and (t_cuts[7]<=Hcal[i]<=t_cuts[8]):
            tau_count+=1
        if (h_cuts[0]<=Ctrk[i]<=h_cuts[1]) and (h_cuts[3]<=Sump[i]<=h_cuts[4]) and (h_cuts[5]<=Ecal[i]<=h_cuts[6]) and (h_cuts[7]<=Hcal[i]<=h_cuts[8]):
            had_count+=1
        if (m_cuts[0]<=Ctrk[i]<=m_cuts[1]) and (m_cuts[3]<=Sump[i]<=m_cuts[4]) and (m_cuts[5]<=Ecal[i]<=m_cuts[6]) and (m_cuts[7]<=Hcal[i]<=m_cuts[8]):
            m_count+=1
        if (e_cuts[0]<=Ctrk[i]<=e_cuts[1]) and (Sump[i]<=e_cuts[2] or e_cuts[3]<=Sump[i]<=e_cuts[4]) and (e_cuts[5]<=Ecal[i]<=e_cuts[6]) and (e_cuts[7]<=Hcal[i]<=e_cuts[8]):
            e_count+=1
    print('Tau:',tau_count,'Hadronic:',had_count,'Electron:',e_count,'Muon:',m_count)
```
